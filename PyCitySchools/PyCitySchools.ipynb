# PyCity Schools Analysis

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# Assigning path to open CSV files 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# ReadingSchool and Student Data Files and store into Pandas DataFrames named respectively
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combining the data into a single dataset named school_data_compliance  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

## District Summary

```python
# Calculating and displaying the total number of unique schools 
school_count = school_data_complete["school_name"].nunique()
school_count
```

```python
# Calculating and displaying the total number of students in schools 
student_count = school_data_complete['Student ID'].count()
student_count
```

```python
# alculating and displaying the total budget of schools
total_budget = sum(school_data_complete['budget'].unique())
total_budget
```

```python
# Calculating and displaying the average (mean) math score and rounding it to 2 decimal places.
average_math_score = round(school_data_complete['math_score'].mean(),2)
average_math_score
```

```python
# Calculating and displaying the average (mean) reading score and rounding it to 2 decimal places.
average_reading_score = round(school_data_complete['reading_score'].mean(),2)
average_reading_score
```

```python
# Calculating and displaying the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = round(passing_math_count / float(student_count) * 100,2)
passing_math_percentage
```

```python
# Calculating and displaying the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete['reading_score'])>=70].count()['student_name']
passing_reading_percentage = round(passing_reading_count/float(student_count)*100,2)
passing_reading_percentage
```

```python
# Calculating and displaying the percentage of students that passed both in math and in reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = round(passing_math_reading_count /  float(student_count) * 100,2)
overall_passing_rate
```

```python
# A high-level snapshot of the district's key metrics in a DataFrame
district_stat = {
    'Total schools' : school_count,
    'Total students': '{:,}'.format(student_count),
    'Total Budget' :'${:,.2f}'.format(total_budget),
    'Average Math Score':average_math_score,
    'Average Reading Score':average_reading_score,
    '% Passing Math':passing_math_percentage,
    '% Passing Reading':passing_reading_percentage,
    '% Overall Passing rate':overall_passing_rate
}
district_summary = pd.DataFrame(district_stat,index=[0])
# Display the DataFrame
district_summary
```

## School Summary

```python
# Calculating and displaying the total students count per school
group_school = school_data_complete.groupby(["school_name"])
per_school_counts = group_school.size()
per_school_counts
```

```python
# Showing the classification of schools as per their type
school_types = group_school['type'].first()
school_types
```

```python
# Calculating and displaying the total school budget and per capita spending per school
per_school_budget = group_school['budget'].first()
print(per_school_budget)
per_school_capita =  per_school_budget/per_school_counts
print(per_school_capita)
```

```python
# Calculating and displaying the average test scores per school both in math and reading and rounding it to 2 decimal places.
per_school_math = round(group_school['math_score'].mean(),2)
print(per_school_math)
per_school_reading = round(group_school['reading_score'].mean(),2)
print(per_school_reading)
```

```python
# Calculating and displaying the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby('school_name').size()
school_students_passing_math = round((students_passing_math/ per_school_counts)*100,2)
school_students_passing_math
```

```python
# CCalculating and displaying the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70].groupby('school_name').size()
school_students_passing_reading = round((students_passing_reading/per_school_counts )*100,2)
school_students_passing_reading
```

```python
# Calculating and displaying the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading_t = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
].groupby('school_name').size()
school_students_passing_math_and_reading = round(students_passing_math_and_reading_t/per_school_counts*100,2)
school_students_passing_math_and_reading
```

```python
# Creating a DataFrame called `per_school_summary` with columns for the calculations that have been done above
school_summary = {'School Type': school_types,
    'Total Students':per_school_counts,
    'Total School Budget':per_school_budget,
    'Per Student Budget':per_school_capita,
    'Average Math Score': per_school_math,
    'Average Reading Score': per_school_reading,
    '% Passing Math': school_students_passing_math,
    '% Passing Reading': school_students_passing_reading,
    '% Overall Passing Rate': school_students_passing_math_and_reading}

per_school_summary = pd.DataFrame(school_summary)

#Formatting 2 columns total school budget and per student budget to approperiate currency format
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sorting the schools by `% Overall Passing` in descending order and display the top 5 rows - top 5 schools with highest passing percentage
top_schools = per_school_summary.sort_values(['% Overall Passing Rate'] ,ascending= False)
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sortingthe schools by `% Overall Passing` in ascending order and display the top 5 rows. - 5 schools with the lowest overall passing percentage
bottom_schools = per_school_summary.sort_values(['% Overall Passing Rate'],ascending= True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
# calculating the scores for different grades
ninth_graders = school_data_complete.loc[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete.loc[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete.loc[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete.loc[(school_data_complete["grade"] == "12th")]

# Determining the average `math_score` of each school for grades 9 ,10 ,11 and 12
ninth_grader_math_scores = ninth_graders.groupby('school_name')['math_score'].mean()
tenth_grader_math_scores = tenth_graders.groupby('school_name')['math_score'].mean()
eleventh_grader_math_scores = eleventh_graders.groupby('school_name')['math_score'].mean()
twelfth_grader_math_scores = twelfth_graders.groupby('school_name')['math_score'].mean()

# Combining each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
   '9th' : round(ninth_grader_math_scores,2),
   '10th' : round(tenth_grader_math_scores,2),
  '11th' : round(eleventh_grader_math_scores,2),
   '12th' : round(twelfth_grader_math_scores,2),
})


# Minor data wrangling
math_scores_by_grade = math_scores_by_grade[['9th', '10th', '11th', '12th']]
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# calculating the scores for different grades
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# GDetermining the average `reading score` of each school for grades 9 ,10 ,11 and 12
ninth_grader_reading_scores = ninth_graders.groupby('school_name')['reading_score'].mean()
tenth_grader_reading_scores = tenth_graders.groupby('school_name')['reading_score'].mean()
eleventh_grader_reading_scores = eleventh_graders.groupby('school_name')['reading_score'].mean()
twelfth_grader_reading_scores = twelfth_graders.groupby('school_name')['reading_score'].mean()

# Combining each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({
   '9th' : round(ninth_grader_reading_scores,2),
   '10th' : round(tenth_grader_reading_scores,2),
  '11th' : round(eleventh_grader_reading_scores,2),
   '12th' : round(twelfth_grader_reading_scores,2)
})

# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Creating a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
# categorizing spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"]= pd.cut(per_school_capita, spending_bins ,labels = spending_labels,include_lowest = True)
school_spending_df.head()
```

```python
#  Calculating the averages for the desired columns for Math , Reading scores with their respective passing percentages and overall Passing rate. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing Rate"].mean()
```

```python
# Assembling data into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": round(spending_math_scores,2),
    "Average Reading Score": round(spending_reading_scores,2),
    "% Passing Math": round(spending_passing_math,2),
    "% Passing Reading": round(spending_passing_reading,2),
    "% Overall Passing Rate": round(overall_passing_spending,2)
})

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins(size of school based on number of students)
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorizing the spending based on the bins(size of school based on number of students)

per_school_summary["School Size"] = pd.cut(per_school_summary['Total Students'],size_bins,labels = labels, include_lowest= True)
per_school_summary
```

```python
# Calculating the averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing Rate"].mean()
```

```python
# Creating a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
size_summary = pd.DataFrame({
    "Average Math Score": round(size_math_scores,2),
    "Average Reading Score": round(size_reading_scores,2),
    "% Passing Math": round(size_passing_math,2),
    "% Passing Reading": round(size_passing_reading,2),
    "% Overall Passing Rate": round(size_overall_passing,2)
})

# Display results
size_summary
```
